# Concepts

- Why linear algebra?
    - Solve big sets of equations
    - Discover structure of data
        - Data compression
- What is a vector?
    - Generalization of an "arrow" with magnitude and direction
    - An object for which addition and scalar multiplication is defined
- What is a vector space?
    - Collection of vectors closed under addition and scalar multiplication
- What is a matrix?
    - Collection of vectors
    - A linear operator
- Matrix invariants
    - Trace
    - Determinant
- Types of matrices (structure)
    - Square
    - Symmetric
    - Identity
    - Diagonal
    - Orthogonal
    - Positive definite
- Eigenvalues and eigenvectors
    - Geometric interpretation
- Matrix multiplication
    - Vector $\times$ vector
    - Matrix $\times$ vector
    - Vector $\times$ matrix
    - Matrix $\times$ matrix
- Matrix multiplication as a linear transform
- Types of matrices (function)
    - Scaling
    - Rotation
    - Reflection
    - Projection
    - Permutation
- Linear independence
    - Linear combinations
    - Span, basis, rank
- Change of basis
    - Coordinate free linear algebra
- Basic operations
    - Gaussian elimination
        - Fundamental subspaces - $C(A), N(A), C(A^T), N(A^T)$
        - LU decomposition
        - LDU decomposition
        - Cholesky decomposition
    - Orthogonalization
        - Gram-Schmidt
        - QR decomposition
    - Projection 
        - Least squares
        - Normal equations
        - PCA
    - Change of basis
        - Spectral decomposition $V\Lambda V^{-1}$
    - SVD
        - Role in PCA
        - Role in solving sets of equations
            - Generalized pseudoinverse
        - Low rank
