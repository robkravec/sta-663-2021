Using R magic in Jupyter, load the `iris` data set from R into a `pandas` DataFrame and use `seaborn` to generate the plot shown.

![img](./iris.png)

```python
iris = %R iris
```

```python
iris.head()
```

```python
import seaborn as sns
sns.set_context('notebook', font_scale=1.5)
```

```python
iris.columns = iris.columns.str.replace('\.Length', 'L').str.replace('\.Width', 'W')
```

```python
g = sns.catplot(
    data=iris.melt(id_vars='Species', var_name='Measurement', value_name='Value'),
    x='Measurement',
    y='Value',
    col='Species',
    kind='violin'
)
g.set_titles('{col_name}')
pass
```

```python
g.savefig('iris.png')
```
